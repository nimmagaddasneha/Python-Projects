```python
print("3) Answer\n \n")
pd.concat([Flower, pd.DataFrame(columns = [ 'Petal ratio', 'Sepal ratio'])], sort = False)
Flower['Petal ratio'] = Flower['Petal.Length']/Flower['Petal.Width']
Flower['Sepal ratio'] = Flower['Sepal.Length']/Flower['Sepal.Width']

Flower

```

```python
print("4 ) Answer")
Petal_mean = Flower['Petal ratio'].mean()
Sepal_mean = Flower['Sepal ratio'].mean()
print(Petal_mean,Sepal_mean)

print("\n\n")

Petal_median = Flower['Petal ratio'].median()
Sepal_median = Flower['Sepal ratio'].median()
print(Petal_median,Sepal_median)


print("\n\n")

Petal_min = Flower['Petal ratio'].min()
Sepal_min = Flower['Sepal ratio'].min()
print(Petal_min,Sepal_min)


print("\n\n")

Petal_max = Flower['Petal ratio'].max()
Sepal_max = Flower['Sepal ratio'].max()
print(Petal_max,Sepal_max)



print("\n\n")

Petal_std = Flower['Petal ratio'].std()
Sepal_std = Flower['Sepal ratio'].std()
print(Petal_std,Sepal_std)

```
