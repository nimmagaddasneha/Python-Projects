### B) Central Limit Theorem:  

The central limit theorem indicates that if the number of samples increase for a distribution, then the sample mean for every sampled population will tend towards a normal distribution. 

For a distribution if the samples increase, there is a high possibility that the sample mean will have a normal curve. Also, if the average of all sample means are taken, it is called a population mean. 

Central limit theorem helps in finding out the proper characteristics of the population, like population mean, population standard deviation, population variance and the population average. 

As the samples increase, the predicted values will be more accurate.
